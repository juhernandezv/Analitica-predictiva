```r
data <- read.csv("housing.csv",sep=";", stringsAsFactors = FALSE)
conserva <- c( "AGE","MEDV")
data <- data[conserva]

MinMaxScaler <- 
function(x) {
    return ((x - min(x)) / (max(x) - min(x)))
}

##
## se aplica la función a los datos
##
data <- as.data.frame(lapply(data, MinMaxScaler))
head(data)
```

```r
distancias1<-dist(data,method="manhattan")
cluster1<-hclust(distancias1)
distancias2<-dist(data,method="euclidean")
cluster2<-hclust(distancias2)
distancias3<-dist(data,method="maximum")
cluster3<-hclust(distancias3)
distancias4<-dist(data,method="canberra")
cluster4<-hclust(distancias4)
```
